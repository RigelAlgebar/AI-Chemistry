<font size='6' style="color:#00A6D6">  <b>3. What is RDkit?</b> </font>
<a name="loading"></a>

[[ Back to Top ]](#content)



RDKit is a collection of cheminformatics and machine-learning software written in C++ and Python.

https://github.com/rdkit/rdkit

<br>

It was developed and used at Rational Discovery since 2000 for building predictive models for ADME, Tox, biological activity. In 2006 it became open source and its development continues with contributions from Novartis.

Documentation is found at https://www.rdkit.org/docs/index.html

RDkit provides several functions to represent molecules in SMILES.

smiles.jpg

This is a one dimensional representation, will also allow to extract several <b>properties</b> already available from RDkit library.

```python
!pip install rdkit-pypi

import rdkit
import rdkit.Chem.Descriptors

#SMILE exaxmple: citral
smile = 'CC(C)=CCCC(C)=CC(=O)'

# Convert into molecular object
mol = rdkit.Chem.MolFromSmiles(smile)

# Extraction of Molecular Weight
MW = rdkit.Chem.Descriptors.MolWt(mol)

print(MW)
```
